<h5 style = "color:red">Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable</h5>

```python
# Choosen variable for which  to calculate summary statistics
chosen_variable = 'salary_in_usd'

# Calculate summary statistics
summary_statistics = {
    'Sum': data[chosen_variable].sum(),
    'Mean': data[chosen_variable].mean(),
    'Standard Deviation': data[chosen_variable].std(),
    'Skewness': data[chosen_variable].skew(),
    'Kurtosis': data[chosen_variable].kurtosis()
}

# Print summary statistics
print("Summary Statistics for Variable '{}'".format(chosen_variable))
for stat, value in summary_statistics.items():
    print(f"{stat}: {value}")
```

<h5 style = "color:red">Write a python program to find out top 15 jobs. Make a bar graph of sales as well</h5>

```python
# Top 15 jobs and bar graph
top_jobs = data['job_title'].value_counts().head(15)
top_15_job_bar_graph = top_jobs.plot(kind='bar', xlabel='Job Title', ylabel='No. of Jobs', title='Top 15 Jobs')
plt.xticks(rotation=45, ha='right')

# Add the value of each bar just above the bar
for bar in top_15_job_bar_graph.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height()}', ha='center', va='bottom')
plt.show()
```

<h5 style = "color:red">Which job has the highest salaries? Illustrate with bar graph.</h5>

```python
# Find the top 5 highest paying jobs
top_5_jobs = data.groupby('job_title')['salary_in_usd'].mean().nlargest(5)
```

```python
# Plotting the bar graph
plt.figure(figsize=(10, 6))
bars = top_5_jobs.plot(kind='bar', xlabel='Job Title', ylabel='Average Salary (USD)', title='Highest Paying Jobs')
plt.xticks(rotation=45, ha='right')

# Add the value of each bar just above the bar
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.2f}', ha='center', va='bottom')
    
plt.tight_layout()
```

<h5 style = "color:red">Write a python program to find out salaries based on experience level. Illustrate it through bar graph.</h5>

```python
# Salaries based on experience level and bar graph
salary_by_experience = data.groupby('experience_level')['salary_in_usd'].mean()
salary_by_experience_bars = salary_by_experience.plot(kind='bar', xlabel='Experience Level', ylabel='Average Salary (USD)', title='Salaries by Experience Level')
plt.xticks(rotation=45, ha='right')

# Add the value of each bar just above the bar
for bar in salary_by_experience_bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.2f}', ha='center', va='bottom')

plt.show()
```

<h5 style = "color:red">Write a Python program to show histogram and box plot of any chosen different variables. Use proper labels in the graph.</h5>

```python
# Choose the variable for which you want to create plots
chosen_column_histogram = 'salary_in_usd'
chosen_column_box_plot = 'salary_in_usd'
salaries = data[chosen_column_histogram]

# Create histogram
num_bins = 15

# Calculate the bin width
bin_width = (salaries.max() - salaries.min()) / (num_bins - 1)

# Define the bin edges
bin_edges = np.arange(salaries.min(), salaries.max() + bin_width, bin_width)

plt.figure(figsize=(10, 6))
counts, bins, _ = plt.hist(data[chosen_column_histogram], bins=bin_edges, color='skyblue', edgecolor='black')
plt.title('Histogram of {}'.format(chosen_column_histogram))
plt.xlabel(chosen_column_histogram)
plt.ylabel('Frequency')

# Set x-axis ticks to match the bin edges
plt.xticks(bin_edges, rotation=45)

# Add bin values just above each bin
for count, bin_edge in zip(counts, bin_edges[:-1]):
    plt.text(bin_edge + bin_width / 2, count, str(int(count)),
             ha='center', va='bottom')
plt.show()

# # Create box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data[chosen_column_box_plot], patch_artist=True, notch=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of {}'.format(chosen_column_box_plot))
plt.xlabel(chosen_column_box_plot)
plt.grid(True)
plt.show()
```
